```python
# D MODELE DE REGRESSION 

# 14. Étudiez comment happiness dépend de la région et des autres variables. 

# Etude de la dépendance entre Happiness et la Region:
# Test d'Anova:

from statsmodels.formula.api import ols

# Ordinary Least Squares (OLS) model
model = ols('Happiness ~Region', data=data).fit()
anova_table = sm.stats.anova_lm(model)
anova_table

# La P-value 5,88e-22 est inférieure à 5% ce qui montre qu'il y a une dépendance entre Happiness et Region.
```
